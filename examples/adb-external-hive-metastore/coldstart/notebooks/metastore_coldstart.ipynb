```python
%sh
export TARGET_HIVE_VERSION="3.1.0"
export TARGET_HADOOP_VERSION="2.7.2"
export TARGET_HIVE_HOME="/opt/apache-hive-${TARGET_HIVE_VERSION}-bin"
export TARGET_HADOOP_HOME="/opt/hadoop-${TARGET_HADOOP_VERSION}"
export JARS_DIRECTORY="/dbfs/tmp/hive/3-1-0/lib/"


if [ ! -d  "$TARGET_HADOOP_HOME" ]; then
if [ ! -f hadoop-${TARGET_HADOOP_VERSION}.tar.gz ]; then
wget https://archive.apache.org/dist/hadoop/common/hadoop-${TARGET_HADOOP_VERSION}/hadoop-${TARGET_HADOOP_VERSION}.tar.gz
fi
tar -xvzf hadoop-${TARGET_HADOOP_VERSION}.tar.gz --directory /opt
fi
if [ ! -d "$TARGET_HIVE_HOME" ]; then
if [ ! -f  apache-hive-${TARGET_HIVE_VERSION}-bin.tar.gz ]; then
wget https://archive.apache.org/dist/hive/hive-${TARGET_HIVE_VERSION}/apache-hive-${TARGET_HIVE_VERSION}-bin.tar.gz
fi
tar -xvzf apache-hive-${TARGET_HIVE_VERSION}-bin.tar.gz --directory /opt
fi
```
